```python
import pandas as pd
from googleapiclient.discovery import build

API_KEY = 'AIzaSyBjIOUeYa_jFDqBWiXtqTlBPPwNcoVc2rc'
```

```python
def get_trending_videos(api_key, max_results = 200):
    youtube = build('youtube', 'v3', developerKey = api_key)
    videos = []  # intialize the list to hold video details
    request = youtube.videos().list(
        part = 'snippet,contentDetails,statistics',
        chart = 'mostPopular',
        regionCode = 'US',
        maxResults = 50
    )
    # paginate through the results if max_results > 50
    while request and len(videos) < max_results:
        response = request.execute()
        for item in response['items']:
            video_details = {
                'video_id': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'published_at': item['snippet']['publishedAt'],
                'channel_id': item['snippet']['channelId'],
                'channel_title': item['snippet']['channelTitle'],
                'category_id': item['snippet']['categoryId'],
                'tags': item['snippet'].get('tags', []),
                'duration': item['contentDetails']['duration'],
                'definition': item['contentDetails']['definition'],
                'caption': item['contentDetails'].get('caption', 'false'),
                'view_count': item['statistics'].get('viewCount', 0),
                'like_count': item['statistics'].get('likeCount', 0),
                'dislike_count': item['statistics'].get('dislikeCount', 0),
                'favorite_count': item['statistics'].get('favoriteCount', 0),
                'comment_count': item['statistics'].get('commentCount', 0)
            }
            videos.append(video_details)
        request = youtube.videos().list_next(request, response)
    return videos[:max_results]
```

```python
def save_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, index = False)

def main():
    trending_videos = get_trending_videos(API_KEY)
    filename = 'yt_trending_videos.csv'
    save_to_csv(trending_videos, filename)
    print(f'Trending videos saved to {filename}')

if __name__ == '__main__':
    main()
```

Collecting the category names to analyze the categories of trending videos:

```python
# API_KEY = 'AIzaSyBjIOUeYa_jFDqBWiXtqTlBPPwNcoVc2rc'
youtube = build('youtube', 'v3', developerKey = API_KEY)
def get_category_mapping():
    request = youtube.videoCategories().list(
        part = 'snippet',
        regionCode = 'US'
    )
    response = request.execute()
    category_mapping = {}
    for item in response['items']:
        category_id = int(item['id'])
        category_name = item['snippet']['title']
        category_mapping[category_id] = category_name
    return category_mapping

category_mapping = get_category_mapping()
category_mapping
```

```python
df['category_name'] = df['category_id'].map(category_mapping)
df['category_name']
```

Now let's analyze the content and duration of videos.
First we need to convert ISO 8601 format of duration to seconds.

```python
!pip install isodate
import isodate
```

```python
df['duration_seconds'] = df['duration'].apply(lambda x: isodate.parse_duration(x).total_seconds())
df['duration_range'] = pd.cut(df['duration_seconds'], bins = [0,300,600, 1200, 3600, 7200], labels = ['0-5 mins', '5-10 mins', '10-20 mins', '20-60 mins', '60-120 mins'])
```

Here, the videos are organized into various duration categories. This allows us to examine and compare engagement metrics based on these time intervals, offering insights into how the length of a video affects viewer behavior and overall video performance.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'duration_seconds', y = 'view_count', data = df, alpha = 0.5)
plt.xlim(0,6000)
plt.title('Video Length vs View Count')
plt.xlabel('Video Length (seconds)')
plt.ylabel('View Count')
plt.tight_layout()
```


This scatter plot shows a slight negative correlation between view count and video length, indicating shorter videos tend to have higher view count.

```python
length_engagement = df.groupby('duration_range')[['view_count','like_count','comment_count']].mean()
length_engagement
```

```python
fig, axes = plt.subplots(1,3,figsize=(18,10))

sns.barplot(y= length_engagement.index, x = length_engagement['view_count'], hue = length_engagement.index, ax = axes[0], palette = 'viridis')
axes[0].set_title('Average View Count by Duration')
axes[0].set_xlabel('Average View Count')
axes[0].set_ylabel('Duration Range')

sns.barplot(y= length_engagement.index, x = length_engagement['like_count'],hue = length_engagement.index, ax = axes[1], palette = 'viridis')
axes[1].set_title('Average Like Count by Duration')
axes[1].set_xlabel('Average Like Count')


sns.barplot(y= length_engagement.index, x = length_engagement['comment_count'],hue = length_engagement.index, ax = axes[2], palette = 'viridis')
axes[2].set_title('Average Comment Count by Duration')
axes[2].set_xlabel('Average Comment Count')


plt.tight_layout()
```

Videos in the range 20-60 mins and 0-5 mins have highest average views,likes and comments, i.e. views either prefer very short length videos or moderate duration videos.

#### What are the peak viewing times for different types of content?

```python
df.head(3)
```

```python
# extracting the hour of posting
df['publish_hr'] = df['published_at'].dt.hour
```

```python
# publish hour distribution
plt.figure(figsize=(12,8))
sns.countplot(x = 'publish_hr', data = df, hue='publish_hr', palette= 'coolwarm')
plt.title('Distribution of Videos by Publishing Hour')
plt.xlabel('Publish Hour')
plt.ylabel('Number of Videos')
```

This distribution shows that most videos are published 15:00 hrs and 19:00 hrs(3 PM - 7 PM), especially around 17:00 hrs. Also a good number of trending videos are also posted around 10 PM. 

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'publish_hr', y ='view_count', data = df, alpha = 0.5, color= 'green')
plt.title('Publishing Hour vs View Count')
plt.xlabel('Publish Hour')
plt.ylabel('View Count')
```

There is a very weak negative correlation between view count and publish hour, indicating that hour of publication has minimal effect on engagement metrics.

## Conclusion:

- The most optimal time for posting videos is around 3 PM - 7 PM.
- The findings indicate that videos with durations of either 0-5 minutes or 20-60 minutes tend to attract the highest levels of viewer engagement.
